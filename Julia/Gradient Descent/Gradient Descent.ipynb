## Gradient Descent

Gradient Descent is a optimization problem to find the local minimum of a function. We will use a process of steps to navigate ourway to the minimum value. 

For this model, we can select any given point on the graph to apply the gradient ascent ablity to the function. Selecting a random point on our graph will allow us to find the tangent line to our function that will return to us the new location of our ascent. This process will occur until the minimum is reached. 
___
Learning Rate: This is the value that the gradient will update at. A higher value will result in over shooting the minimim but a lower value will return us a small update. Finding the blance between these values will return to us the minimum at a tangable rate.

____

Today, We will attempt to find the local minimum of a function.

```julia
f(x) = (x+4)^2

∇f(x) = 2*(x+4)
plot(f, 
    xlim = (-10,2),ylim = (-2,25),
    legend = false)
```

The typical gradient descent function will consist of the following update function.

##### <center> X<sub>K+1</sub> = X<sub>K</sub> - $\alpha$ * $\triangledown$ f(X<sub>k</sub>) </center>


Each of the following variables are as of:
<li> X<sub>K+1</sub> : new updated point in our function </li>
<li> X<sub>K</sub> : The last point in our graph that had placed. </li>
<li> $\alpha$ : Step size or learning rate of the descent. </li>
<li> $\triangledown$ f(X<sub>k</sub>) : Derivative function </li>


Below, we pick a random point on our graph as our start to find the minimum. 

The function we are using today is:

##### <center> f(x) = (x+4)<sup>2</sup> </center>

Where the derivative of this function is:


##### <center> f(x)<sup>-1</sup> = 2 * (x+4) </center>


The supposed minimum of this function is -4. Let's see if we can get the value with the gradient model. 
____

```julia
#get the value for the minimum 

derivative(f,-4)

#We know that the minimum is -4, so this should return 0
```

We can now create a function to give us the output of results that we need for gradient descent and we can see the output of each descent until we reach a significant point in the udpate process. 
___

```julia
x7
```

```julia

f(x) = (x)^2
∇f(x) = 2*(x)

function ascent(fun,der,point, α, ϵ)
    
    """ One function to output full gradient descent update in a graph
    
    Args:
        fun ([equation/function]): Input of the function we want to find the minimum of in the gradient.
    
        def ([equation/function]): Input of the derivative of the function that we want the minimum of.
    
        point (int):   Guess of the first point in the function. You enter a starting point that you start your ascent of the gradient from.
    
        α (float):     Step size of our descent. 
    
        ϵ (float):     The value in the diffrence between udpates that is significant in our model.
        
    
    Output:
        A list of the update X-values that we are at on each update of the gradient & A plot of the gradient model. 
    
    
    """
    plot(f,legend = false)
    scatter!([(point,f(point))])
    x0= point
    x1 = point
    i = 1
    println("Gradient update  |    X-value: ")
    
    while abs(f(x1)) > ϵ
        x0 = x1
        x1 = x0 - α*∇f(x0)
        scatter!([(x1,f(x1))])
        println("       " , i , "         |     ", x1)
        
        
        i += 1
    end
    println("Final Update value:", x1 )
    plot!()
    
end
```

```julia
ascent(f, ∇f, 4,.3,.00001)
```

___
Above we see the output of the plot after the gradient descent. We can see the updated points are going towards the minimum value in our graph. This graph stops updating after 8 timmes achiving the minimum point in our graph.

#### Application of Gradient Descent on to linear regression

In regression, we want to fit a line that fits all the points that we have in the data. We can adjust weights in a linear formula to best fit the outputs that we aim to have in the model. We can combine the use of a basic linear regression line and the means squared error cost function. 

#### Mean Square cost function: MSE(W,b) = $\frac{1}{N}\sum$ $\frac{dL}{dW}$(y - (w<sub>x</sub> - b) )<sup>2</sup>

We can find the derivative for the cost function with respect to w and b. This will result into 2 function to update.

#### $\triangledown \frac{dc}{dw}\sum$  = $\frac{2}{N}\sum$ (y<sub>i</sub> - (w<sub>x</sub> - b) )* x<sub>i</sub>

#### $\triangledown \frac{dc}{db}\sum$  = $\frac{-1}{N}\sum$ (y<sub>i</sub> - w<sub>x</sub> - b)


When we update our weights and bias or intercept, we will use this update function

#### w<sup>k+1</sup> = w<sup>k</sup> - $\alpha \triangledown \frac{dc}{dw}$(w<sup>x</sup>,b<sup>2</sup>)

#### b<sup>k+1</sup> = b<sup>k</sup> - $\alpha \triangledown \frac{dc}{db}$(w<sup>x</sup>,b<sup>2</sup>)

#### Below, we use linear regression data and apply the gradient descent model to it. 

___

```julia
#data = [(2,2.2), (3,3.75), (4.5,3.5), (5.2,4.5),(6.2,5),(7.2,6.5),(3.4,4.5),(6.6,5.3), (3,5), (4.5,5.5), (3.5,3), (4.4,5), (5,5.5)]
data = [(2,2.2), (3,3.75), (4.5,3.5), (5.2,4.5),(6.2,5),(7.2,6.5),(3.4,4.5),(6.6,5.3), (3,5), (4.5,5.5), (3.5,3), (4.4,5), (5,5.5)]
scatter(data)
```

```julia
C(w) = .5* sum([(w[1]*x[1] + w[2])^2 for x in data])
∇C(w) = 1/4*[sum([(w[1]*x[1] + w[2] - x[2]) * x[1] for x in data]), sum([(w[1]*x[1] + w[2] - x[2]) for x in data])]
```

```julia
w = randn(2)
for i = 1:1000
    w = w -.01 * ∇C(w)
end
```

```julia
scatter(data, legend = false)
plot!(x -> x*w[1] + w[2],
    xlim = (1,10))
```

___
Here we use gradient descent to get a linear regression line for the data. We updated our model 1000 times over gradient descent with an alpha step of .01. This model is useful to get the linear model but the value of the step size and times we update really is important. If we pick the wrong values we will over shoot the values and the weights will not produce a result as a line. 

Gradient descent and mean square error can combine to produce a regression line.
