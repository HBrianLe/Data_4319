# Decison Trees

In Machine Learning we have the issue of decison analysis that can be used to show represent decison making. With this, we can use what are decison trees models to do tasks such as regression and classification. Decison Trees are a powerful tool that can be used to quickly classify data in a if or else format of a tree with decision rules.
___
Decison Tress are build off of the idea of an upside down tree where the root is at the top or start and many other branches and leaves that are attached to it. There are many types of trees such as a binary tree where each node or branch is a refrence to two or none other nodes. 

Trees are a a very simple model that involves both real world applications and machine learning. In the real world, we use the same topic of decison trees to make our sisuation and spots we are placed in. Similar to machine learning, it uses the data to determine classifications on data or on the features it has to work with. In a typical tree, the classification criteria is with a less than or greater than value of a feature that can signal a diffrent classification or other methods such as a yes or no binary classification. 

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
```

### Wine Dataset
Today, we will be looking at a Wine Type classification dataset that can be found [here](https://archive.ics.uci.edu/ml/datasets/wine). This dataset has three types of wine classes and 13 features that make up each type. Each type is distinctively diffrent from each other and that we can use tress to serpate these classifications,

```python
wine = pd.read_csv("wines.csv")
wine
```

```python
#selcting the data between features and labels. 
label = wine["Class"]
features = wine.loc[:, wine.columns != 'Class']
```

```python
#below, we use the decison tree function that is built in SKlearn to build the Model. We give it the max depth for how many layers the tree should have to avoid overfitting. 
tree_clf = DecisionTreeClassifier(max_depth= 3, max_features= 4)
#we use the fit fucntion to place the feature and labels to run from. As the output it produces a tree model
tree_clf.fit(features, label)
```

```python
tree_clf.predict(features)
```

___
### Overfitting

In decison trees, we face the issue of overfitting. This is where and the point where we overclassifiy data in the tree into many other branches that are unnecessary for model prediction. Although it can improve the output of the model now and the data that we have, the result is a overfit or too tuned model will not help in predicting models well in the future of unknown data. 

With less data or outfit in our model, we can use it to predict with a larger range of that is unknown like for instnace an outlier in the data that can be misclassified if too fine tuned and be placed in the wrong classifiation.

## Train and test Split

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.30)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_clf.fit(X_train, y_train)
```

```python
predictions = tree_clf.predict(X_test)
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

# Random Forest 

If you have a group of decison tress classifiers, you create a Random Forest that is an ensemble of them. This is a classification or regression model that outputs the mean or avearge of the production of other individual trees. The use of random forest allows for the minimum of the overfitting issue that decison trees produce. 

We will use the decison tree model as before but we will use the random forest model instead. 
___

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(features, label)
```

```python
rfc_pred = rfc.predict(features)
```

```python
print(confusion_matrix(label,rfc_pred))
```

```python
rfc.predict(features)
```

When we look at the output of the Random Forest, we see the confusion matrix produces results that do not have any errors in the classifications. Unlike the Decison Tree that did have errors, this does not. Not only does the forest look better than trees, it is much more powerful to produce better results.

#### Advantages and Disadvantages

**Advantages**:
- *Very simple to setup and understand the output.*

**Disadvantages**:
- *You can overcomplex a model that is due to overfitting and faces issues of future data.*

- *Hard to know where the best fit of the paremeters and depth of the tree for best output.*
